1. Explain the differences between AI, ML, Deep Learning (DL),Data

-   Artificial Intelligence (AI): The broadest concept, AI is the
    effort to automate intellectual tasks normally performed by humans. It's the umbrella term for creating systems that can perform tasks that typically require human intelligence, such as learning, problem-solving, perception, and decision-making.

    Machine Learning (ML): A subset of AI, ML is a set of techniques that allow computers to learn from data without being explicitly programmed. Instead of hard-coding rules, ML algorithms use statistical methods to enable machines to "learn" from data and make predictions or decisions.

    Deep Learning (DL): A subset of ML, DL is a type of ML that uses artificial neural networks with multiple layers (hence "deep"). These networks are inspired by the structure and function of the human brain and are particularly effective for tasks involving complex patterns in data, such as image and speech recognition.

    Data Science (DS): An interdisciplinary field that uses scientific methods, processes, algorithms, and systems to extract knowledge and insights from structured and unstructured data. Data Science encompasses various techniques, including ML, statistics, data mining, and data visualization, to analyze and interpret data for decision-making.

2.What are the types of machine learning? Describe each with one
  real-world example.

-   Supervised Learning: This type of ML uses labeled data to train
    algorithms. The algorithm learns to map input data to output labels.

    Example: Spam detection in email. The algorithm is trained on a dataset of emails labeled as "spam" or "not spam" and learns to classify new emails accordingly.
    Unsupervised Learning: This type of ML uses unlabeled data to find hidden patterns or structures within the data. There are no predefined output labels.

    Example: Customer segmentation. An algorithm can analyze customer purchasing behavior data to group customers into different segments based on their similarities without being told what those segments should be.
    Reinforcement Learning: This type of ML involves an agent learning to make decisions by taking actions in an environment to maximize a reward. The agent learns through trial and error.

    Example: Training a robot to walk. The robot receives a reward for taking steps and reaching a destination and a penalty for falling, learning the optimal movements through this feedback.
    Semi-Supervised Learning: This approach combines a small amount of labeled data with a large amount of unlabeled data during training. It's useful when obtaining large amounts of labeled data is difficult or expensive.

    Example: Image recognition with a limited labeled dataset. A model can use the labeled images to learn basic features and then use the much larger set of unlabeled images to further refine its understanding and improve accuracy.
    Self-Supervised Learning: A form of unsupervised learning where the data itself provides the supervision. The model learns by predicting parts of the input from other parts of the input.

    Example: Predicting the next word in a sentence. A language model can be trained on a large text corpus to predict the next word based on the preceding words, effectively creating its own learning task from the data.

3.  Define overfitting, underfitting, and the bias-variance    
    tradeoff in machine learning.
-   Overfitting: This occurs when a model learns the training data
    too well, including the noise and outliers. As a result, the model performs poorly on unseen data because it doesn't generalize well. It's like memorizing the answers to a test instead of understanding the concepts.

    Underfitting: This happens when a model is too simple to capture the underlying patterns in the data. It performs poorly on both the training data and unseen data because it hasn't learned the relationships effectively. It's like not studying enough for a test and not being able to answer even the basic questions.

    Bias-Variance Tradeoff: This is a fundamental concept in machine learning.

    Bias: The error introduced by approximating a real-world problem, which may be complicated, by a simplified model. High bias suggests the model has made strong assumptions about the data and is too simple.
    Variance: The error introduced by the model's sensitivity to small fluctuations in the training data. High variance suggests the model is too complex and is capturing noise in the training data.
    The tradeoff is that reducing bias often increases variance, and vice versa. The goal is to find a balance between bias and variance to build a model that generalizes well to unseen data. A good model has low bias and low variance.

4. What are outliers in a dataset, and list three common techniques
   for handling them.

-   Outliers: Data points that significantly deviate from other
    observations in a dataset. They can occur due to errors in data collection, natural variations, or unusual events. Outliers can negatively impact the performance of machine learning models by distorting statistical measures and model training.

    Techniques for handling outliers:

    Removal: The simplest approach is to remove the outlier data points from the dataset. This should be done cautiously, especially if the dataset is small, as it can lead to loss of valuable information. It's most suitable when outliers are clearly due to errors.
    
    Transformation: Applying mathematical transformations (e.g., log transformation, square root transformation) to the data can reduce the impact of outliers by compressing the range of values.
    
    Imputation: Replacing outliers with a more representative value, such as the mean, median, or a value estimated by a model. This approach retains the data points but modifies their values to be less extreme.
    
    Capping/Winsorizing: Limiting the values of outliers to a certain threshold. This involves setting a maximum or minimum value, and any data point beyond this threshold is replaced with the threshold value.

5.  Explain the process of handling missing values and mention one
    imputation technique for numerical and one for categorical data.

-   Handling Missing Values: Missing values are common inreal-world
    datasets and can lead to biased results or errors in machine learning models. The process of handling them typically involves:

    Identification: Detecting which values are missing in the dataset.
    
    Understanding the Cause: Investigating why the data is missing (e.g., data entry errors, sensor malfunction, data not applicable). This understanding helps in choosing the appropriate handling technique.
    Deciding on a Strategy: Choosing a method to address the missing data. Common strategies include:
    
    Deletion: Removing rows or columns with missing values. This is suitable if only a small percentage of data is missing and deleting doesn't lead to significant information loss.
    Imputation: Replacing missing values with estimated values. This is often preferred when deletion would lead to a substantial loss of data.
    Keeping Missing Values: Some models can handle missing values directly.
    
    Imputation Techniques:
    For Numerical Data: Mean, Median, or Mode Imputation. Replacing missing numerical values with the mean, median, or mode of the non-missing values in that column.
    Example: If a dataset has a missing value in a 'Age' column, you could replace it with the average age of all other individuals in the dataset.
    For Categorical Data: Mode Imputation. Replacing missing categorical values with the mode (most frequent category) of the non-missing values in that column.
    Example: If a dataset has a missing value in a 'City' column, you could replace it with the city that appears most frequently in the dataset.

8. Write a Python script to:
    ● Generate 1000 samples from a normal distribution.
    ● Introduce 50 random missing values.
    ● Fill missing values with the column mean.
    ● Plot a histogram before and after imputation.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate 1000 samples from a normal distribution
np.random.seed(42) # for reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)

# 2. Introduce 50 random missing values
missing_indices = np.random.choice(1000, size=50, replace=False)
data_with_missing = data.copy()
data_with_missing[missing_indices] = np.nan

# 3. Fill missing values with the column mean
mean_value = np.nanmean(data_with_missing)
data_imputed = np.nan_to_num(data_with_missing, nan=mean_value)

# 4. Plot histograms before and after imputation
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data_with_missing, bins=30, edgecolor='black')
plt.title('Histogram Before Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(data_imputed, bins=30, edgecolor='black')
plt.title('Histogram After Imputation (Mean)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Address outliers

### Subtask:
Detect and handle outliers in the transaction amount using a chosen technique (e.g., capping, transformation, or removal if appropriate).


**Reasoning**:
Define a function for capping outliers using the IQR method, calculate the IQR and bounds for 'TransactionAmount', apply the capping function, and visualize the distribution after capping using histogram and box plot.



```python
# 1. Define a function to cap outliers using the Interquartile Range (IQR) method.
def cap_outliers_iqr(series, lower_bound_multiplier=1.5, upper_bound_multiplier=1.5):
    """
    Caps outliers in a pandas Series using the IQR method.

    Args:
        series (pd.Series): The input Series with potential outliers.
        lower_bound_multiplier (float): The multiplier for the IQR to determine the lower bound.
        upper_bound_multiplier (float): The multiplier for the IQR to determine the upper bound.

    Returns:
        pd.Series: The Series with outliers capped.
    """
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - lower_bound_multiplier * IQR
    upper_bound = Q3 + upper_bound_multiplier * IQR

    # Cap values outside the bounds
    capped_series = series.copy()
    capped_series[capped_series < lower_bound] = lower_bound
    capped_series[capped_series > upper_bound] = upper_bound

    return capped_series

# 2. Calculate the first quartile (Q1), third quartile (Q3), and the IQR for the 'TransactionAmount' column.
Q1_amount = df['TransactionAmount'].quantile(0.25)
Q3_amount = df['TransactionAmount'].quantile(0.75)
IQR_amount = Q3_amount - Q1_amount

print(f"Q1 (TransactionAmount): {Q1_amount}")
print(f"Q3 (TransactionAmount): {Q3_amount}")
print(f"IQR (TransactionAmount): {IQR_amount}")

# 3. Calculate the lower and upper bounds for outlier detection using the IQR method (e.g., lower_bound = Q1 - 1.5 * IQR, upper_bound = Q3 + 1.5 * IQR).
lower_bound_amount = Q1_amount - 1.5 * IQR_amount
upper_bound_amount = Q3_amount + 1.5 * IQR_amount

print(f"Lower Bound (TransactionAmount): {lower_bound_amount}")
print(f"Upper Bound (TransactionAmount): {upper_bound_amount}")


# 4. Apply the capping function to the 'TransactionAmount' column of the DataFrame df using the calculated bounds.
df['TransactionAmount_Capped'] = cap_outliers_iqr(df['TransactionAmount'])

# 5. Visualize the distribution of the 'TransactionAmount' column after capping using a histogram and a box plot to observe the effect of outlier handling.
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(df['TransactionAmount_Capped'], bins=50, edgecolor='black')
plt.title('Distribution of Transaction Amount After Capping')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
df.boxplot(column='TransactionAmount_Capped')
plt.title('Box Plot of Transaction Amount After Capping')
plt.ylabel('Transaction Amount')

plt.tight_layout()
plt.show()
```
